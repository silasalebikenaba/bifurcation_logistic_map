```python
import numpy as np
import matplotlib.pyplot as plt

# Inputs
r = float(input("Choose r, the bifurcation parameter, in (0,4): "))
n = int(input("Choose n, the number of iterations: "))
x = np.zeros(n + 1)
x[0] = float(input("Choose x(1), the initial point, in [0,1]: "))

# Plot y = x
s = np.arange(0, 1.01, 0.2)
plt.plot(s, s, 'b')

# Plot logistic map
t = np.arange(0, 1.001, 0.01)
v = r * t * (1 - t)
plt.plot(t, v, 'g')

# Compute iterations
for i in range(1, n + 1):
    x[i] = r * x[i - 1] * (1 - x[i - 1])

# Cobwebbing
plt.plot([x[0], x[0]], [x[0], x[1]], 'k')
for i in range(1, n):
    plt.plot([x[i - 1], x[i]], [x[i], x[i]], 'k')
    plt.plot([x[i], x[i]], [x[i], x[i + 1]], 'k')

plt.xlabel("x_n")
plt.ylabel("x_{n+1}")
plt.title("Cobwebbing Diagram for Logistic Map")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Inputs
rmin = float(input("Input the minimum value of r: "))
rmax = float(input("Input the maximum value of r: "))

n = 850     # number of r values
N = 100     # number of iterations

plt.figure(figsize=(8, 6))

for m in range(1, n + 1):
    r = rmin + (rmax - rmin) * m / n
    x = 0.5

    # Transient iterations
    for _ in range(N):
        x = r * x * (1 - x)

    # Store points
    X = np.zeros(N + 1)
    R = np.zeros(N + 1)
    X[0] = x
    R[0] = r

    for k in range(N):
        X[k + 1] = r * X[k] * (1 - X[k])
        R[k + 1] = r

    plt.plot(R, X, ',k')   # ',' = pixel marker

plt.xlabel("r")
plt.ylabel("x")
plt.title("Bifurcation Diagram of the Logistic Map")
plt.show()
```
